### Добавление целочисленных категорий наличия и количества детей

```python
def children_category(row):
    if row['children'] == 0:
        return 0
    if row['children'] == 1:
        return 1
    return 2
   
df['children_cat'] = df.apply(children_category, axis=1)
df.groupby('children_cat')['children'].unique()
```

### Добавление целочисленных категорий дохода

```python
display(df['total_income'].describe())
low = df['total_income'].describe()[4]
medium = low + (df['total_income'].describe()[6] - low)*2

def total_income_category(row):
    if pd.isna(row['total_income']):
        return row['total_income']
    if row['total_income'] <= low:
        return 0
    if row['total_income'] <= medium:
        return 1
    return 2
   
df['total_income_cat'] = df.apply(total_income_category, axis=1)
df['total_income_cat'].astype('int64', copy=False, errors='ignore')
display(df.groupby('total_income_cat')['total_income'].describe().T)
```

### Добавление целочисленных категорий типов занятости

```python
display(df['income_type'].value_counts(dropna=False))

def income_type_category(row):
    if row['income_type'] == 'госслужащий':
        return 3
    if row['income_type'] == 'пенсионер':
        return 2
    if row['income_type'] == 'компаньон':
        return 1
    return 0
   
df['income_type_cat'] = df.apply(income_type_category, axis=1)
display(df.groupby('income_type_cat')['income_type'].unique())
```

### Добавление целочисленных категорий целей кредита

```python
frame_for_lemmas = df['purpose'].value_counts(dropna=False).to_frame().reset_index()

m = Mystem()

def lemmatisizer(row):
    return ' '.join(m.lemmatize(row['index'])) 
 
frame_for_lemmas['lemmatisized'] = frame_for_lemmas.apply(lemmatisizer, axis = 1)
frame_for_lemmas
```

```python
def purpose_category(row):
    if 'автомобиль' in row['lemmatisized']:
        return 2
    if 'свадьба' in row['lemmatisized']:
        return 0
    if 'образование' in row['lemmatisized']:
        return 3
    return 1

frame_for_lemmas['purpose_cat'] = frame_for_lemmas.apply(purpose_category, axis = 1)

frame_for_lemmas
```

```python
purpose_category_dict = frame_for_lemmas.groupby('index')['purpose_cat'].mean().to_dict()

def purpose_category_for_df(row):
    return purpose_category_dict[row['purpose']]

df['purpose_cat'] = df.apply(purpose_category_for_df, axis=1)
df.groupby('purpose_cat')['purpose'].unique().to_frame()
```

### Добавление целочисленных категорий пола

```python
display(df['gender'].value_counts(dropna=False))

def gender_category(row):
    if 'F' in row['gender']:
        return 0
    return 1

df['gender_cat'] = df.apply(gender_category, axis = 1)

display(df['gender_cat'].value_counts(dropna=False))
```

### Заполнение нулевых значений возраста

```python
def find_corr_for_col(col):
    display(df.corr()[col].abs().sort_values(ascending=False))
```

```python
find_corr_for_col('dob_years')
```

```python
display(df['dob_years'].describe())

years_dict = df.groupby(['income_type_cat', 'children_cat', 'gender_cat'])['dob_years'].median().to_dict()

def fill_zero_age(row):
    if row['dob_years'] == 0:
        return years_dict.get((row['income_type_cat'], row['children_cat'], row['gender_cat']))
    return row['dob_years']

df['dob_years'] = df.apply(fill_zero_age, axis = 1)

display(df['dob_years'].describe())
```

### Заполнение пропущенных значений трудового стажа

```python
find_corr_for_col('days_employed')
```

```python
display(df['days_employed'].describe())

days_dict = df.groupby(['dob_years', 'income_type_cat'])['days_employed'].median().to_dict()

def fill_na_days(row):
    if pd.isna(row['days_employed']):
        return days_dict.get((row['dob_years'], row['income_type_cat']))
    return row['days_employed']

df['days_employed'] = df.apply(fill_na_days, axis = 1)

display(df['days_employed'].describe())
```

### Заполнение пропущенных значений дохода

```python
find_corr_for_col('total_income')
```

```python
display(df['total_income'].describe())

income_dict = df.groupby(['gender_cat', 'education_id'])['total_income'].median().to_dict()

def fill_na_income(row):
    if pd.isna(row['total_income']):
        return income_dict.get((row['gender_cat'], row['education_id']))
    return row['total_income']

df['total_income'] = df.apply(fill_na_income, axis = 1)

display(df['total_income'].describe())
```

### Итоговая проверка и корректировка пропусков/дубликатов/типов данных

```python
df.info()
```

```python
def fill_na_total_income_category(row):
    if pd.isna(row['total_income_cat']):
        if pd.isna(row['total_income']):
            return row['total_income']
        if row['total_income'] <= low:
            return 0
        if row['total_income'] <= medium:
            return 1
        return 2
    return row['total_income_cat']
   
df['total_income_cat'] = df.apply(fill_na_total_income_category, axis=1)
display(df.groupby('total_income_cat')['total_income'].describe().T)
```

## Ответы на вопросы

### Есть ли зависимость между наличием детей и возвратом кредита в срок?

```python
children_pivot = df.pivot_table(index='children_cat', values='debt', aggfunc='mean')
children_pivot.rename(columns={'debt':'Вероятность просрочки'}, inplace=True)
children_pivot = children_pivot.T
children_pivot.rename(columns={0:'Нет детей', 1:'Один ребёнок', 2:'Два и более'}, inplace=True)
children_pivot
```

Зависимость между наличием детей и возвратом кредита наблюдается. Бездетные реже имеют просрочки по кредитам - 7,5% против 9,1% у заёмщиков с одним ребёнком и 9,3% с двумя и более.

### Есть ли зависимость между семейным положением и возвратом кредита в срок?

```python
family_pivot = (
    df
    .pivot_table(index='family_status', values='debt', aggfunc='mean')
    .sort_values('debt', ascending=True)
)
family_pivot.rename(columns={'debt':'Вероятность просрочки'}, inplace=True)
family_pivot = family_pivot.T
family_pivot
```

Зависимость между семейным положением и возвратом кредита наблюдается. Вдовцы и вдовы, а так же разведённые реже имеют просрочки по кредитам - 6,6% и 7,1% соответственно.
