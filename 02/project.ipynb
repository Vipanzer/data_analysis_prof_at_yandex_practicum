## Добавление параметров и подготовка дата сета к анализу<a id='step_3'></a>

- Цена квадратного метра

```python
df['price_per_sq_m'] = df['last_price'] / df['total_area']
```

```python
df['price_per_sq_m'] = df['price_per_sq_m'].round().astype('int64')
```

- День недели, месяц и год публикации объявления

```python
df['weekday'] = df['first_day_exposition'].dt.weekday + 1
df['month'] = df['first_day_exposition'].dt.month
df['year'] = df['first_day_exposition'].dt.year
```

- Категоризация по этажу квартиры: первый, последний, другой

```python
def floor_category(row):
    if row['floor'] == 1:
        return 1
    if row['floor'] == row['floors_total']:
        return 3
    return 2

df['floor_category'] = df.apply(floor_category, axis=1)
```

- Соотношения жилой к общей площади, и площади кухни к общей.

```python
columns = ['total_area', 'last_price', 'price_per_sq_m', 'rooms', 'ceiling_height']

def create_cell(shell, text):
    payload = dict(
        source='set_next_input',
        text=text,
        replace=False,
    )
    shell.payload_manager.write_payload(payload, single=False)

def investigate_column(column):
    print(column.describe(percentiles=[.05, .25, .75, .95]))
    print(column.value_counts())

def create_hist(column):
    df.plot(y = column, kind = 'hist', bins = 200, grid=True, figsize = (12,6))

def create_clear_hist(column):
    Q1 =  df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    ranges = (0, Q3+3*IQR)
    df.plot(y = column, kind = 'hist', bins = 200, grid=True, figsize = (12,6), range = ranges)

def generate_column_exploratory_description(df):
    shell = get_ipython()
    for column in reversed(columns):
        create_cell(shell, dedent(
            '''...'''
        ))
        create_cell(shell, dedent(
            f'''create_clear_hist("{column}")'''
        ))
        create_cell(shell, dedent(
            f'''Гистограмма распределения квартир по "{column}".'''
        ))
        create_cell(shell, dedent(
            f'''create_hist("{column}")'''
        ))
        create_cell(shell, dedent(
            f'''ninetynineth = np.percentile(df["{column}"], 99)
df[df["{column}"] > ninetynineth].head(15)'''
        ))
        create_cell(shell, dedent(
            f'''first = np.percentile(df["{column}"], 1)
df[df["{column}"] < first].head(15)'''
        ))
        create_cell(shell, dedent(
            f'''investigate_column(df["{column}"])'''
        ))
        create_cell(shell, dedent(
            f'''- Изучение "{column}"'''
        ))
```

Явно выделяются типовые значения. Значения в 2 метра и ниже похожи на ошибки в введённых данных, заменим их медианными значениями по этажности домов. Значения более четырёх вполне объяснимых метров немного сложнее. До 6 метров включительно, вероятно, специфические предложения. От 7 до 20 включительно и 100 - ошибки, которые можно заменить медианными значениями для домов схожей этажности. Значения больше 20 разделим на 10.

```python
df.loc[(df['ceiling_height'] > 20) & (df['ceiling_height'] < 99), 'ceiling_height'] = (
    df.loc[(df['ceiling_height'] > 20) & (df['ceiling_height'] < 99), 'ceiling_height'] / 10
)
```

```python
median_ceiling_height_by_floors_total = df.groupby('floors_total')['ceiling_height'].median().to_dict()

def ceiling_height_correction(row):
    if row['ceiling_height'] <= 2:
        return median_ceiling_height_by_floors_total[row['floors_total']]
    if row['ceiling_height'] >= 7:
        return median_ceiling_height_by_floors_total[row['floors_total']]
    return row['ceiling_height']

df['ceiling_height'] = df.apply(ceiling_height_correction, axis=1)
```
