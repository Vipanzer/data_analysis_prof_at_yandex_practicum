    - Avg_class_frequency_current_month — средняя частота посещений в неделю за предыдущий месяц;
    - Avg_additional_charges_total — суммарная выручка от других услуг фитнес-центра: кафе, спорт-товары, косметический и массажный салон.

## Исследовательский анализ данных

```python
#Функция анализа признаков
def eda_for_features(feature):
#Таблица describe признака со столбцами для всего датасета и с разбивкой по факту оттока
    describe_frame_for_feature = (
        df[feature].describe().to_frame().reset_index()
        .merge(df.loc[df['churn'] == 0][feature].describe().to_frame().reset_index(), on='index')
        .merge(df.loc[df['churn'] == 1][feature].describe().to_frame().reset_index(), on='index')
    )
    describe_frame_for_feature.columns = [feature, 'all', 'churn = 0', 'churn = 1']
    display(describe_frame_for_feature)
#Таблица value_counts признака со столбцами для всего датасета и с разбивкой по факту оттока
    if df[feature].nunique() < 40:
        values_share_for_feature = (
            df[feature].value_counts(normalize=True).to_frame().reset_index()
            .merge(df.loc[df['churn'] == 0][feature].value_counts(normalize=True).to_frame().reset_index(), on='index')
            .merge(df.loc[df['churn'] == 1][feature].value_counts(normalize=True).to_frame().reset_index(), on='index')
        )
        values_share_for_feature.columns = [feature, 'all', 'churn = 0', 'churn = 1']
        display(values_share_for_feature)
#Таблица среднего показателя оттока для разных групп по признаку
    if df[feature].nunique() < 20:
        display(df.groupby(feature)['churn'].mean().to_frame())
#Столбчатая гистограмма признака со столбцами для всего датасета и с разбивкой по факту оттока
    fig, ax = plt.subplots(figsize=(16, 10))
    ax1 = sns.histplot(data=df, x=feature, stat='probability', kde=True, element="step", color='green', alpha=0.1)
    ax2 = sns.histplot(data=df, x=feature, hue='churn', stat='probability', kde=True, element="step", alpha=0.1)
    plt.ylabel('Вероятность', fontsize=18)
    plt.xlabel(feature, fontsize=18)
    ax.grid()
    plt.show()
```

```python
#Функция генерации ячеек
def create_cell(shell, text):
    payload = dict(
        source='set_next_input',
        text=text,
        replace=False,
    )
    shell.payload_manager.write_payload(payload, single=False)

def generate_feature_exploratory_description(features):
    shell = get_ipython()
    for feature in reversed(features):
        create_cell(shell, dedent('''#### Вывод
 - 
 - 
 - '''))
        create_cell(shell, dedent(f'''eda_for_features("{feature}")'''))
        create_cell(shell, dedent(f'''### Изучение "{feature}"'''))
```

```python
# убираю по одному из двух пар сильно скоррелированных признаков
df_ml = df.drop(['month_to_end_contract', 'avg_class_frequency_total'], axis=1)

# разделяю данные на признаки (матрица X) и целевую переменную (y)
X = df_ml.drop(columns=['churn'])
y = df_ml['churn']

# разделяю модель на обучающую и валидационную выборку
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# стандартизирую обучающую и валидационную матрицы признаков
scaler = StandardScaler()
X_train_st = scaler.fit_transform(X_train)
X_test_st = scaler.transform(X_test)
```

```python
print('LogisticRegression')
model_1 = LogisticRegression(solver='liblinear', random_state=0)
model_1.fit(X_train_st, y_train)
y_pred_1 = model_1.predict(X_test_st)
y_prob_1 = model_1.predict_proba(X_test_st)[:,1]
print('Accuracy: {:.2f}'.format(accuracy_score(y_test, y_pred_1)))
print('Precision: {:.2f}'.format(precision_score(y_test, y_pred_1)))
print('Recall: {:.2f}'.format(recall_score(y_test, y_pred_1)))
print('F1: {:.2f}'.format(f1_score(y_test, y_pred_1)))
print('ROC_AUC: {:.2f}'.format(roc_auc_score(y_test, y_prob_1)))
```

```python
print('RandomForestClassifier')
model_2 = RandomForestClassifier(random_state=0)
model_2.fit(X_train_st, y_train)
y_pred_2 = model_2.predict(X_test_st)
y_prob_2 = model_2.predict_proba(X_test_st)[:,1]
print('Accuracy: {:.2f}'.format(accuracy_score(y_test, y_pred_2)))
print('Precision: {:.2f}'.format(precision_score(y_test, y_pred_2)))
print('Recall: {:.2f}'.format(recall_score(y_test, y_pred_2)))
print('F1: {:.2f}'.format(f1_score(y_test, y_pred_2)))
print('ROC_AUC: {:.2f}'.format(roc_auc_score(y_test, y_prob_2)))
```

```python
feature_importances = model_2.feature_importances_
fig, ax = plt.subplots(figsize=(16, 10))
sns.barplot(x=feature_importances, y=X.columns.tolist())
plt.show()
```

### Выводы
 - Логистическая регрессия, не смотря на небольшое отставание в точности, опередила Случайный лес по полноте, F1 и площади под кривой ошибок.
