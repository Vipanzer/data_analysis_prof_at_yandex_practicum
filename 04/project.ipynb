```python
def create_cell(shell, text):
    payload = dict(
        source='set_next_input',
        text=text,
        replace=False,
    )
    shell.payload_manager.write_payload(payload, single=False)

def investigate_column(column):
    print(column.describe(percentiles=[.05, .25, .75, .95]))
    print(column.value_counts())

def create_hist(column):
    df.plot(y = column, kind = 'hist', bins = 200, grid=True, figsize = (9,6))
    plt.title('Гистограмма '+ column + "\n")

def create_clear_hist(column):
    Q1 =  df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    ranges = (0, Q3+3*IQR)
    df.plot(y = column, kind = 'hist', bins = 200, grid=True, figsize = (9,6), range = ranges)
    plt.title('Гистограмма без выбросов '+ column + "\n")

def generate_column_exploratory_description(df):
    shell = get_ipython()
    for column in reversed(df.columns):
        create_cell(shell, dedent(
            '''...'''
        ))
        create_cell(shell, dedent(
            f'''create_clear_hist("{column}")'''
        ))
        create_cell(shell, dedent(
            f'''create_hist("{column}")'''
        ))
        create_cell(shell, dedent(
            f'''ninetynineth = np.percentile(df["{column}"], 99)
df[df["{column}"] > ninetynineth].head(15)'''
        ))
        create_cell(shell, dedent(
            f'''first = np.percentile(df["{column}"], 1)
df[df["{column}"] <= first].head(15)'''
        ))
        create_cell(shell, dedent(
            f'''investigate_column(df["{column}"])'''
        ))
        create_cell(shell, dedent(
            f'''- Изучение "{column}"'''
        ))
```

Гистограмма даёт первое общее представление о поколениях игровых платформ. Выделяется "золотой век" с 2005 по 2011.

```python
missing_dates_1 = df.loc[df['year_of_release'].isna()].drop(['genre', 'na_sales', 'eu_sales',
       'jp_sales', 'other_sales', 'critic_score', 'user_score', 'rating'], axis=1)

def digit_from_name(row):
    split_name = row['name'].split()
    if split_name[-1].isdigit():
        res = split_name[-1]
        return res

missing_dates_1['digits_in_name'] = missing_dates_1.apply(digit_from_name, axis=1)
missing_dates_1.sample(10)
```

```python
missing_dates_1 = missing_dates_1.dropna(subset=['digits_in_name'])

def two_or_four_digits(row):
    digits = row['digits_in_name']
    if len(digits) == 2:
        return 1
    if len(digits) == 4:
        return 1
    return 0

missing_dates_1['correct_num_of_digits'] = missing_dates_1.apply(two_or_four_digits, axis=1)

missing_dates_1 = missing_dates_1.loc[missing_dates_1['correct_num_of_digits'] == 1]
missing_dates_1
```

```python
missing_dates_1 = missing_dates_1.drop(['correct_num_of_digits'], axis=1)
missing_dates_1['digits_in_name'] = missing_dates_1['digits_in_name'].astype('int64')

def repair_digits(row):
    if row['digits_in_name'] < 17:
        return row['digits_in_name'] + 2000
    if 80 < row['digits_in_name'] < 100:
        return row['digits_in_name'] + 1900
    return row['digits_in_name']

missing_dates_1['digits_in_name'] = missing_dates_1.apply(repair_digits, axis=1)

missing_dates_1 = missing_dates_1.loc[missing_dates_1['digits_in_name'] > 1979]
missing_dates_1['digits_in_name'] = missing_dates_1['digits_in_name'] - 1
missing_dates_1
```

```python
missing_dates_1_dict = missing_dates_1.groupby(['name', 'platform'])['digits_in_name'].mean().to_dict()

def fill_na_missing_dates_1(row):
    if pd.isna(row['year_of_release']):
        return missing_dates_1_dict.get((row['name'], row['platform']))
    return row['year_of_release']

df['year_of_release'] = df.apply(fill_na_missing_dates_1, axis=1)

df.loc[df['year_of_release'].isna()].sample(10)
```

```python
platform_min_year_dict = df.groupby(['platform'])['year_of_release'].min().to_dict()
platform_min_year_dict
```

```python
platform_min_year_dict = df.groupby(['platform'])['year_of_release'].min().to_dict()
platform_min_year_dict
```

```python
missing_dates_2_names = df.loc[df['year_of_release'].isna()]['name'].value_counts().index.to_list()

missing_dates_2 = (
    df.query('name in @missing_dates_2_names')
    .drop(['na_sales', 'eu_sales', 'jp_sales', 'other_sales', 'critic_score', 'user_score', 'rating'], axis=1)
    .groupby(['name']).agg({'year_of_release': ['unique', 'min', 'max']})
)
missing_dates_2.set_axis(['unique', 'min', 'max'], axis = 'columns', inplace = True)
missing_dates_2 = missing_dates_2.dropna()
missing_dates_2['diff'] = missing_dates_2['max'] -  missing_dates_2['min']
missing_dates_2 = missing_dates_2.sort_values(by='diff', ascending=False)

missing_dates_2_dict = missing_dates_2.query('diff < 2')['min'].to_dict()

def fill_na_missing_dates_2(row):
    if pd.isna(row['year_of_release']):
        if row['platform'] in platform_min_year_dict.keys():
            if row['name'] in missing_dates_2_dict.keys():
                if missing_dates_2_dict.get(row['name']) >= platform_min_year_dict.get(row['platform']):
                    return missing_dates_2_dict.get(row['name'])
    return row['year_of_release']

df['year_of_release'] = df.apply(fill_na_missing_dates_2, axis=1)

df.loc[df['year_of_release'].isna()].sample(10)
```

```python
# эти пять переизданий потребовали индивидуального подхода

missing_dates_3 = missing_dates_2.query('diff > 1')
missing_dates_3_names = missing_dates_3.index.to_list()
display(df.loc[df['year_of_release'].isna()].query('name in @missing_dates_3_names').sort_values(by='name'))

re_release_dict = {'Mario Tennis': 2012, 
                   'Smashing Drive': 2002, 
                   'Sonic the Hedgehog': 2006, 
                   'Space Invaders': 1980, 
                   'Wheel of Fortune': 2003}

def fill_na_missing_dates_3(row):
    if pd.isna(row['year_of_release']):
        if row['name'] in missing_dates_3_names:
            return re_release_dict.get(row['name'])                    
    return row['year_of_release']

df['year_of_release'] = df.apply(fill_na_missing_dates_3, axis=1)

df.loc[df['year_of_release'].isna()].sample(10)
```

```python
platform_median_year_dict = df.groupby(['platform'])['year_of_release'].median().to_dict()

def fill_na_missing_dates_4(row):
    if pd.isna(row['year_of_release']):
        return platform_median_year_dict.get(row['platform'])
    return row['year_of_release']

df['year_of_release'] = df.apply(fill_na_missing_dates_4, axis=1)

df.loc[df['year_of_release'].isna()]
```

## Проверка гипотез

Средние пользовательские рейтинги платформ Xbox One и PC одинаковые.

Н0 - Средние пользовательские рейтинги платформ Xbox One и PC **равны**, т.е. средние значения равны.

Н1 - Средние пользовательские рейтинги платформ Xbox One и PC **не равны**, т.е. средние значения не равны.

Для проверки гипотезы будет использоваться критерий p-value.

```python
df_hypo = df.dropna()
sample_1 = df_hypo.loc[(df_hypo['platform'] == 'XOne') & (df_hypo['year_of_release'] >= 2013)]['user_score'] 
sample_2 = df_hypo.loc[(df_hypo['platform'] == 'PC') & (df_hypo['year_of_release'] >= 2013)]['user_score']

stat, p = levene(sample_1, sample_2)
p
```

```python
results = st.ttest_ind(sample_1, sample_2, equal_var = False)
alpha = 0.05

print('p-value: ', results.pvalue)
if results.pvalue < alpha:
    print('Принимаем альтернативную гипотезу')
else:
    print('Опровергнуть нулевую гипотезу нельзя')
```

Величина p-value не позволяет говорить уверенно об одинаковых средних пользовательских рейтингах платформ Xbox One и PC, но и не позволяет опровергнуть.
