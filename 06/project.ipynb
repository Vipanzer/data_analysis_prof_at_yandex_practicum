## Изучение и подготовка данных

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()

%matplotlib inline

pd.set_option('max_columns', None)
```

```python
visits = pd.read_csv('visits_log.csv')
orders = pd.read_csv('orders_log.csv')
costs = pd.read_csv('costs.csv')
datasets = (visits, orders, costs)
for dataset in datasets:
    display(dataset.sample(10))
    display(dataset.info())
    display(dataset.duplicated().sum())
```

Для анализа предоставлены три датасета с информацией о посещениях сайта Яндекс.Афиши, заказах и рекламных расходах.

**Описание данных:**
1. Посещения:
    - Device — тип устройства;
    - End Ts — время завершения сессии;
    - Source Id — айди источника;
    - Start Ts — время начала сессии;
    - Uid — айди пользователя;


2. Заказы:
    - Buy Ts — время совершения заказа;
    - Revenue — стоимость заказа;
    - Uid — айди пользователя;


3. Расходы:
    - source_id — айди источника;
    - dt — дата;
    - costs — расходы.


- Требуется привести названия столбцов к единому виду - нижний регистр + нижние подчёркивания вместо пробелов.
- Все временные метки  необходимо привести к типу данных datetime и добавить столбцы для категоризации.
- Для удобства работы нужно изменить айди источников на строковые значение типа id1, id2 и т.д.

```python
def dataset_cols_corr(dataset):
    display(dataset.columns)
    dataset.columns = dataset.columns.str.lower()
    dataset.columns = dataset.columns.str.replace(" ", "_")
    display(dataset.columns)

dataset_cols_corr(visits)
dataset_cols_corr(orders)
```

```python
display(visits['end_ts'].describe())
visits['end_ts'] = pd.to_datetime(visits['end_ts'])
display(visits['end_ts'].describe())

display(visits['start_ts'].describe())
visits['start_ts'] = pd.to_datetime(visits['start_ts'])
display(visits['start_ts'].describe())

display(visits.info())

display(orders['buy_ts'].describe())
orders['buy_ts'] = pd.to_datetime(orders['buy_ts'])
display(orders['buy_ts'].describe())

display(orders.info())

display(costs['dt'].describe())
costs['dt'] = pd.to_datetime(costs['dt'])
costs['dt'] = costs['dt'].dt.date.astype('datetime64')
display(costs['dt'].describe())

display(costs.info())
```

```python
visits['session_day'] = visits['start_ts'].dt.date.astype('datetime64')
visits['session_week'] = visits['start_ts'].values.astype('datetime64[W]')
visits['session_month'] =  visits['start_ts'].values.astype('datetime64[M]')
display(visits.info())
display(visits.sample(10))

orders['order_day'] = orders['buy_ts'].dt.date.astype('datetime64')
orders['order_week'] = orders['buy_ts'].values.astype('datetime64[W]')
orders['order_month'] = orders['buy_ts'].values.astype('datetime64[M]')
display(orders.info())
display(orders.sample(10))

costs['spent_week'] = costs['dt'].values.astype('datetime64[W]')
costs['spent_month'] = costs['dt'].values.astype('datetime64[M]')
display(costs.info())
display(costs.sample(10))
```

```python
display(visits['source_id'].unique())
display(visits['source_id'].value_counts())
visits['source_id'] = visits['source_id'].astype(str)
visits['source_id'] = 'id' + visits['source_id']
display(visits['source_id'].unique())
display(visits['source_id'].value_counts())

display(costs['source_id'].unique())
display(costs['source_id'].value_counts())
costs['source_id'] = costs['source_id'].astype(str)
costs['source_id'] = 'id' + costs['source_id']
display(costs['source_id'].unique())
display(costs['source_id'].value_counts())
```

### Вывод
Данные предоставлены в почти готовом для работы виде без пропусков и дубликатов. Потребовалось лишь заменить тип данных в столбцах с временными метками и привести к общему виду названия столбцов, а также для удобства добавить столбцы с элементами дат и привести обозначение айди источников трафика к виду id1, id2 и т.д.

Имеются все необходимые для выполнения поставленных задач данные.

### Среднее количество пользовательских сессий в день

```python
sessions = visits.groupby('session_day')['uid'].agg({'nunique', 'count'}).reset_index()
sessions = sessions.rename(columns={'nunique': 'daily_active_users', 'count': 'daily_sessions'})
sessions['daily_sessions_per_user'] = sessions['daily_sessions'] / sessions['daily_active_users']
sessions_total = sessions['daily_sessions_per_user'].mean()
display(sessions_total.round(3))
```

```python
fig, ax = plt.subplots(figsize=(16, 10))
ax = sns.lineplot(x='session_day', y='daily_sessions_per_user', data=sessions)
ax.set_xlim('2017-06-01', '2018-05-31')
plt.ylabel('Количество сессий', fontsize=18)
plt.xlabel('Дни', fontsize=18)
ax.grid()
plt.title('Среднее количество пользовательских сессий в день за период с июня 2017 по май 2018', fontsize=20)
plt.show()
```

```python
sessions_by_device = visits.groupby(['device', 'session_day'])['uid'].agg({'nunique', 'count'}).reset_index()
sessions_by_device = sessions_by_device.rename(columns={"nunique": "daily_active_users", "count": "daily_sessions"})
sessions_by_device['daily_sessions_per_user'] = sessions_by_device['daily_sessions'] / sessions_by_device['daily_active_users']
sessions_by_device.sample(10)
```

```python
sessions_by_device_total = sessions_by_device.pivot_table(index = 'device', values = 'daily_sessions_per_user', aggfunc = 'mean')
display(sessions_by_device_total.round(3))
```

```python
fig, ax = plt.subplots(figsize=(16, 10))
ax = sns.lineplot(x='session_day', y='daily_sessions_per_user', hue='device', data=sessions_by_device,)
ax.set_xlim('2017-06-01', '2018-05-31')
plt.ylabel('Количество сессий', fontsize=18)
plt.xlabel('Дни', fontsize=18)
ax.grid()
plt.title('Среднее количество пользовательских сессий в день за период с июня 2017 по май 2018', fontsize=20)
plt.show()
```

#### Вывод
- В среднем за рассматриваемый период пользователи заходили на сайт раз в день, и лишь небольшая их часть (8.2%) позволяла себе дополнительную сессию.


- С десктопов (8.9%) заходят на сайт чуть чаще, чем с сенсорных устройств (5.1%).

### Средняя продолжительность пользовательской сессии - ASL

```python
visits.loc[visits['end_ts'] < visits['start_ts']]
```

```python
row_177972_corr_start = visits.loc[177972, 'end_ts']
row_177972_corr_end = visits.loc[177972, 'start_ts']

row_4181_corr_start = visits.loc[4181, 'end_ts']
row_4181_corr_end = visits.loc[4181, 'start_ts']

display(row_177972_corr_start, row_177972_corr_end)
display(row_4181_corr_start, row_4181_corr_end)
```

```python
visits.at[177972, 'start_ts'] = row_177972_corr_start
visits.at[177972, 'end_ts'] = row_177972_corr_end
visits.at[4181, 'start_ts'] = row_4181_corr_start
visits.at[4181, 'end_ts'] = row_4181_corr_end
```

```python
visits['session_duration_sec'] = (visits['end_ts'] - visits['start_ts']).dt.seconds
display(visits['session_duration_sec'].describe())
display(visits['session_duration_sec'].value_counts(sort=True, ascending=False, dropna=False))

zero_duration_part = len(visits[visits['session_duration_sec'] == 0]) / len(visits)
print('Сессии нулевой длительности: {:.2%}'.format(zero_duration_part))
```

```python
over_one_hour_part = len(visits[visits['session_duration_sec'] > 3600]) / len(visits)
print('Сессии дольше часа: {:.2%}'.format(over_one_hour_part))
      
over_two_hours_part = len(visits[visits['session_duration_sec'] > 3600*2]) / len(visits)
print('Сессии дольше двух часов: {:.2%}'.format(over_two_hours_part))
      
over_three_hours_part = len(visits[visits['session_duration_sec'] > 3600*3]) / len(visits)
print('Сессии дольше трёх часов: {:.2%}'.format(over_three_hours_part))
      
display(visits[visits['session_duration_sec'] <= 3600*3]['session_duration_sec'].describe())
display(visits[visits['session_duration_sec'] <= 3600*3]['session_duration_sec'].value_counts(sort=True, ascending=False, dropna=False))
```

### Когортный анализ коэффициента удержания - Retention Rate

```python
first_session_month = visits.groupby(['uid'])['session_month'].min()
first_session_month.name = 'first_session_month'
visits = visits.join(first_session_month, on='uid')
visits['cohort_lifetime'] = visits['session_month'] - visits['first_session_month']
visits['cohort_lifetime'] = visits['cohort_lifetime'] / np.timedelta64(1,'M')
visits['cohort_lifetime'] = visits['cohort_lifetime'].round().astype('int')
visits.head(10)
```

```python
cohorts = visits.groupby(['first_session_month','cohort_lifetime']).agg({'uid':'nunique'}).reset_index()

initial_users_count = cohorts[cohorts['cohort_lifetime'] == 0][['first_session_month','uid']]
initial_users_count = initial_users_count.rename(columns={'uid':'cohort_users'})
cohorts = cohorts.merge(initial_users_count,on='first_session_month')

cohorts.head(15)
```

```python
cohorts['retention'] = cohorts['uid']/cohorts['cohort_users']

retention_pivot = cohorts.pivot_table(index='first_session_month',columns='cohort_lifetime',values='retention',aggfunc='sum')

plt.figure(figsize=(16, 10))
sns.set(style='white')
sns.heatmap(retention_pivot.drop(columns={0}), annot=True, fmt='.1%', linewidths=1, linecolor='gray')
plt.title('Коэффициент удержания по месячным когортам за период с июня 2017 по май 2018', fontsize=20)
```

```python
retention_mean = cohorts.groupby('cohort_lifetime').agg({'retention': 'mean'}).reset_index()

retention_mean.style.format({'retention': '{:.2%}'})
```
