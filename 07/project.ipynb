### Описание данных

```python
display(orders['date'].describe(datetime_is_numeric=True))
orders['date'] = orders['date'].map(lambda x: dt.datetime.strptime(x, '%Y-%m-%d'))
display(orders['date'].describe(datetime_is_numeric=True))

display(visitors['date'].describe(datetime_is_numeric=True))
visitors['date'] = visitors['date'].map(lambda x: dt.datetime.strptime(x, '%Y-%m-%d'))
display(visitors['date'].describe(datetime_is_numeric=True))
```

```python
orders = orders.sort_values(by=['date', 'group'])
```

```python
multi_group_visitors = (
    orders
    .groupby('visitorId')['group'].nunique()
    .sort_values(ascending=False)
    .reset_index()
)
multi_group_visitors = multi_group_visitors[multi_group_visitors['group'] > 1]

print(
    'Доля пользователей вошедших в обе группы {:.2%}'
    .format(multi_group_visitors['visitorId'].nunique()/orders['visitorId'].nunique()))

display(
    orders
    .loc[orders['visitorId'].isin(multi_group_visitors['visitorId'])]
    .pivot_table(index = 'visitorId', columns = 'group', values = 'revenue', aggfunc = ('count', 'sum'))
    .reset_index()
)
```

- **Это говорит о возможных технических ошибках в реализации теста, стоило бы сообщить ответственным.**

```python
visitors_before = orders['visitorId'].nunique()
transactions_before = orders['transactionId'].nunique()
revenue_sum_before = orders['revenue'].sum()
```

```python
orders = orders.loc[~(orders['visitorId'].isin(multi_group_visitors['visitorId']))]
```

```python
visitors_after = orders['visitorId'].nunique()
transactions_after = orders['transactionId'].nunique()
revenue_sum_after = orders['revenue'].sum()

visitors_diff_in_perc = (visitors_before - visitors_after)/visitors_before
transactions_diff_in_perc = (transactions_before - transactions_after)/transactions_before
revenue_sum_diff_in_perc = (revenue_sum_before - revenue_sum_after)/revenue_sum_before


print(
    'До удаления количество пользователей было {:d}, после удаления стало {:d}, уменьшилось на {:.2%}'
    .format(visitors_before, visitors_after, visitors_diff_in_perc))
print(
    'До удаления количество транзакций было {:d}, после удаления стало {:d}, уменьшилось на {:.2%}'
    .format(transactions_before, transactions_after, transactions_diff_in_perc))
print(
    'До удаления общий объём выручки составил {:d}, после удаления стал {:d}, уменьшился на {:.2%}'
    .format(revenue_sum_before, revenue_sum_after, revenue_sum_diff_in_perc))
```

```python
hypothesis['Reach /%median'] = hypothesis['Reach'] / hypothesis['Reach'].median() *100
hypothesis['Impact /%_median'] = hypothesis['Impact'] / hypothesis['Impact'].median() *100
hypothesis['Confidence /%median'] = hypothesis['Confidence'] / hypothesis['Confidence'].median() *100
hypothesis['Efforts /%median'] = hypothesis['Efforts'] / hypothesis['Efforts'].median() *100

hypothesis
```

```python
hypothesis['ICE'] = hypothesis['Impact'] * hypothesis['Confidence'] / hypothesis['Efforts']

ice_top = hypothesis.sort_values(by='ICE', ascending=False)[['Hypothesis', 'ICE']]
ice_top
```

```python
hypothesis['RICE'] = hypothesis['Reach'] *  hypothesis['Impact'] * hypothesis['Confidence'] / hypothesis['Efforts']

rice_top = hypothesis.sort_values(by='RICE', ascending=False)[['Hypothesis', 'RICE']]
rice_top
```

## Анализ результатов A/B-теста

```python
# Подготавливаю датасет с кумулятивными показателями.
# Скелет из дат и групп.
datesGroups = orders[['date','group']].drop_duplicates()

# Нанизываю на скелет данные о заказах, кумулятивные показатели получены с помощью лямбда-функции.
ordersAggregated = datesGroups.apply(
lambda x: orders[np.logical_and(orders['date'] <= x['date'], orders['group'] == x['group'])].agg({
'date' : 'max',
'group' : 'max',
'transactionId' : pd.Series.nunique,
'visitorId' : pd.Series.nunique,
'revenue' : 'sum'}), axis=1).sort_values(by=['date','group'])

# Нанизываю на скелет данные о посетителях.
visitorsAggregated = datesGroups.apply(
lambda x: visitors[np.logical_and(visitors['date'] <= x['date'], visitors['group'] == x['group'])].agg({
'date' : 'max',
'group' : 'max',
'visitors' : 'sum'}), axis=1).sort_values(by=['date','group'])

# Собираю датасет.
cumulativeData = ordersAggregated.merge(visitorsAggregated, left_on=['date', 'group'], right_on=['date', 'group'])
cumulativeData.columns = ['date', 'group', 'orders', 'buyers', 'revenue', 'visitors']

display(cumulativeData)
```

### Кумулятивная выручка

```python
fig, ax = plt.subplots(figsize=(16, 10))
ax = sns.lineplot(x='date', y='revenue', hue='group', data=cumulativeData,)
ax.set_xlim(cumulativeData['date'].min(), cumulativeData['date'].max())
ax.set_ylim(0, 5250000)
plt.xticks(cumulativeData['date'], rotation=90)
plt.yticks(np.arange(0, 5500000, step=250000))
plt.ylabel('Кумулятивная выручка', fontsize=18)
plt.xlabel('Дни', fontsize=18)
ax.grid()
plt.title('График кумулятивной выручки по группам', fontsize=20)
plt.show()
```

#### Вывод
Выручка почти равномерно увеличивается в течение всего теста, за исключением гигантского скачка на 19 день в группе B. Это может быть как всплеск числа заказов, так и очень дорогой единичный заказ.

### Кумулятивный средний чек

```python
cumulativeData['avg_check'] = cumulativeData['revenue']/cumulativeData['orders']

fig, ax = plt.subplots(figsize=(16, 10))
ax = sns.lineplot(x='date', y='avg_check', hue='group', data=cumulativeData,)
ax.set_xlim(cumulativeData['date'].min(), cumulativeData['date'].max())
ax.set_ylim(3500, 10500)
plt.xticks(cumulativeData['date'], rotation=90)
plt.yticks(np.arange(3500, 10750, step=250))
plt.ylabel('Кумулятивный средний чек', fontsize=18)
plt.xlabel('Дни', fontsize=18)
ax.grid()
plt.title('График кумулятивного среднего чека по группам', fontsize=20)
plt.show()
```

#### Вывод
Средний чек начинает стабилизироваться уже с середины теста, за исключением гигантского скачка на 19 день и последовавшей за ним коррекцией в группе B. Преположение о всплеске числа заказов можно отбросить, это похоже очень дорогой единичный заказ.

### Относительное изменение кумулятивного среднего чека группы B к группе A

```python
avg_check_pivot = cumulativeData.pivot_table(index = 'date', columns = 'group', values = 'avg_check').reset_index()
avg_check_pivot['B_to_A_ratio'] = avg_check_pivot['B'] / avg_check_pivot['A'] - 1

fig, ax = plt.subplots(figsize=(16, 10))
ax = sns.lineplot(x='date', y='B_to_A_ratio', data=avg_check_pivot,)
ax.set_xlim(avg_check_pivot['date'].min(), avg_check_pivot['date'].max())
ax.set_ylim(-0.55, 0.55)
plt.xticks(avg_check_pivot['date'], rotation=90)
plt.yticks(np.arange(-0.55, 0.55, step=0.05))
plt.xlabel('Дни', fontsize=18)
plt.axhline(y=0, color='black', linestyle='--')
ax.grid()
plt.title('График относительного изменения кумулятивного среднего чека группы B к группе A', fontsize=20)
plt.show()
```

#### Вывод
После скачков в первой половине теста с 15 дня наблюдается стабилизация, которую нарушает резкий скачок в 19 день. Последующая постепенная коррекция подтверждает стабилизацию.

### Кумулятивная конверсия

```python
cumulativeData['conversion'] = cumulativeData['orders']/cumulativeData['visitors']

fig, ax = plt.subplots(figsize=(16, 10))
ax = sns.lineplot(x='date', y='conversion', hue='group', data=cumulativeData,)
ax.set_xlim(cumulativeData['date'].min(), cumulativeData['date'].max())
ax.set_ylim(0.02, 0.037)
plt.xticks(cumulativeData['date'], rotation=90)
plt.yticks(np.arange(0, 0.04, step=0.001))
plt.ylabel('Кумулятивная конверсия', fontsize=18)
plt.xlabel('Дни', fontsize=18)
ax.grid()
plt.title('График кумулятивной конверсии по группам', fontsize=20)
plt.show()
```

### Статистическая значимость различий в конверсии по сырым данным

```python
# Собираю датасет для оценки статистической значимости.
# Ежедневные посетители в группе A.
visitorsADaily = visitors[visitors['group']=='A'][['date', 'visitors']]
visitorsADaily.columns = ['date', 'visitorsPerDateA']

# Кумулятивные посетители в группе A.
visitorsACummulative = visitorsADaily.apply(lambda x: visitorsADaily[visitorsADaily['date'] <= x['date']].agg({'date' : 'max', 'visitorsPerDateA' : 'sum'}), axis=1)
visitorsACummulative.columns = ['date', 'visitorsCummulativeA']

# Ежедневные посетители в группе B.
visitorsBDaily = visitors[visitors['group']=='B'][['date', 'visitors']]
visitorsBDaily.columns = ['date', 'visitorsPerDateB']

# Кумулятивные посетители в группе B.
visitorsBCummulative = visitorsBDaily.apply(lambda x: visitorsBDaily[visitorsBDaily['date'] <= x['date']].agg({'date' : 'max', 'visitorsPerDateB' : 'sum'}), axis=1)
visitorsBCummulative.columns = ['date', 'visitorsCummulativeB']


# Ежедневные количество и сумма заказов в группе A.
ordersADaily = orders[orders['group']=='A'][['date', 'transactionId', 'visitorId', 'revenue']]\
    .groupby('date', as_index=False)\
    .agg({'transactionId' : pd.Series.nunique, 'revenue' : 'sum'})
ordersADaily.columns = ['date', 'ordersPerDateA', 'revenuePerDateA']

# Кумулятивные количество и сумма заказов в группе A.
ordersACummulative = ordersADaily.apply(
    lambda x: \
        ordersADaily[ordersADaily['date'] <= x['date']]\
            .agg({
                'date' : 'max',
                'ordersPerDateA' : 'sum',
                'revenuePerDateA' : 'sum'}), axis=1).sort_values(by=['date'])
ordersACummulative.columns = ['date', 'ordersCummulativeA', 'revenueCummulativeA']

# Ежедневные количество и сумма заказов в группе B.
ordersBDaily = orders[orders['group']=='B'][['date', 'transactionId', 'visitorId', 'revenue']]\
    .groupby('date', as_index=False)\
    .agg({'transactionId' : pd.Series.nunique, 'revenue' : 'sum'})
ordersBDaily.columns = ['date', 'ordersPerDateB', 'revenuePerDateB']

# Кумулятивные количество и сумма заказов в группе B.
ordersBCummulative = ordersBDaily.apply(
    lambda x: \
        ordersBDaily[ordersBDaily['date'] <= x['date']]\
            .agg({
                'date' : 'max',
                'ordersPerDateB' : 'sum',
                'revenuePerDateB' : 'sum'}), axis=1).sort_values(by=['date'])
ordersBCummulative.columns = ['date', 'ordersCummulativeB', 'revenueCummulativeB']

# Объединяю.
data = ordersADaily.merge(ordersBDaily, left_on='date', right_on='date', how='left')\
    .merge(ordersACummulative, left_on='date', right_on='date', how='left')\
    .merge(ordersBCummulative, left_on='date', right_on='date', how='left')\
    .merge(visitorsADaily, left_on='date', right_on='date', how='left')\
    .merge(visitorsBDaily, left_on='date', right_on='date', how='left')\
    .merge(visitorsACummulative, left_on='date', right_on='date', how='left')\
    .merge(visitorsBCummulative, left_on='date', right_on='date', how='left')

data.head(5)
```

**Н0** - Статистически значимых различий в конверсии **нет**.

**Н1** - Статистически значимые различия в конверсии **есть**.

```python
# Готовлю данные о посетителях сделавших заказы по группам.
ordersByUsersA = orders[orders['group']=='A'].groupby('visitorId', as_index=False).agg({'transactionId' : pd.Series.nunique})
ordersByUsersA.columns = ['visitorId', 'orders']

ordersByUsersB = orders[orders['group']=='B'].groupby('visitorId', as_index=False).agg({'transactionId' : pd.Series.nunique})
ordersByUsersB.columns = ['visitorId', 'orders'] 

# В сэмплах для стат теста объединяю их с посетителями без заказов в виде серии нулей
# в количестве разницы между общим числом посетителей в группе и покупателями.
sampleA = pd.concat(
    [ordersByUsersA['orders'], \
     pd.Series(0, index=np.arange(data['visitorsPerDateA'].sum() - len(ordersByUsersA['orders'])), name='orders')],axis=0)

sampleB = pd.concat(
    [ordersByUsersB['orders'], \
     pd.Series(0, index=np.arange(data['visitorsPerDateB'].sum() - len(ordersByUsersB['orders'])), name='orders')],axis=0)

print('p-value: {0:.3f}'.format(st.mannwhitneyu(sampleA, sampleB, alternative = 'two-sided')[1]))

print('Относительный прирост B к A: {:.0%}'.format(sampleB.mean()/sampleA.mean()-1))
```

#### Вывод
- P-value значительно меньше 0.05, что говорит о наличии статистически значимых различий в конверсии между группами.
- Относительный прирост конверсии группы B к конверсии группы A - 16%.

### Статистическая значимость различий в конверсии по очищенным данным

```python
# Готовлю фильтры.
usersWithManyOrders = pd.concat([ordersByUsersA[ordersByUsersA['orders'] > 1]['visitorId'], ordersByUsersB[ordersByUsersB['orders'] > 1]['visitorId']], axis = 0)
usersWithExpensiveOrders = orders[orders['revenue'] > 55000]['visitorId']
abnormalUsers = pd.concat([usersWithManyOrders, usersWithExpensiveOrders], axis = 0).drop_duplicates().sort_values()
display(abnormalUsers.head(5))
display(abnormalUsers.shape)
```

```python
# Готовлю фильрованные сэмплы для теста.
sampleAFiltered = pd.concat(
    [ordersByUsersA[np.logical_not(ordersByUsersA['visitorId'].isin(abnormalUsers))]['orders'], \
     pd.Series(0, index=np.arange(data['visitorsPerDateA'].sum() - len(ordersByUsersA['orders'])),name='orders')],axis=0)

sampleBFiltered = pd.concat(
    [ordersByUsersB[np.logical_not(ordersByUsersB['visitorId'].isin(abnormalUsers))]['orders'], \
     pd.Series(0, index=np.arange(data['visitorsPerDateB'].sum() - len(ordersByUsersB['orders'])),name='orders')],axis=0)
```

**Н0** - Статистически значимых различий в конверсии **нет**.

**Н1** - Статистически значимые различия в конверсии **есть**.

```python
print('p-value: {0:.3f}'.format(st.mannwhitneyu(sampleAFiltered, sampleBFiltered)[1], alternative = 'two-sided'))
print('Относительный прирост B к A: {:.0%}'.format(sampleBFiltered.mean()/sampleAFiltered.mean()-1))
```

#### Вывод
- P-value уменьшилось почти в три раза после очистки, что подтверждает предыдущий вывод о наличии статистически значимых различий в конверсии между группами.
- Относительный прирост конверсии группы B к конверсии группы A увеличился с 16% до 19%.
