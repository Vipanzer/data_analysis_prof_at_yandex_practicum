## Выполнение поставленных задач

### Посчитать, сколько книг вышло после 1 января 2000 года

```python
query1 = ''' 
SELECT
    COUNT(*) AS cnt,
    COUNT(DISTINCT book_id) AS unq_book_ids,
    COUNT(DISTINCT title) AS unq_titles
FROM
    books
WHERE
    publication_date > '2000-01-01';
'''

result1 = pd.io.sql.read_sql(query1, con = engine)

result1
```

#### Вывод
- После 1 января 2000 года вышло 818 книг.
- В таблице есть книга переизданная.

### Для каждой книги посчитать количество обзоров и среднюю оценку

```python
query2 = ''' 
SELECT
    DISTINCT books.title AS title,
    SUM(SUBQ1.reviews_cnt) AS reviews_cnt,
    AVG(SUBQ2.avg_rating) AS avg_rating
FROM
    books
    LEFT JOIN (
        SELECT
            book_id AS book_id,
            COUNT(review_id) AS reviews_cnt
        FROM
            reviews
        GROUP BY
            book_id) AS SUBQ1 ON books.book_id = SUBQ1.book_id
    LEFT JOIN (
        SELECT
            book_id AS book_id,
            AVG(rating) AS avg_rating
        FROM
            ratings
        GROUP BY
            book_id) AS SUBQ2 ON books.book_id = SUBQ2.book_id
GROUP BY
    title;
'''

result2 = pd.io.sql.read_sql(query2, con = engine)

result2.sample(15)
```

### Определить издательство, которое выпустило наибольшее число книг толще 50 страниц

```python
query3 = ''' 
SELECT
    publishers.publisher AS publisher,
    SUBQ1.books_cnt AS books_cnt
FROM
    publishers
    RIGHT JOIN (
        SELECT
            publisher_id AS publisher_id,
            COUNT(book_id) AS books_cnt
        FROM
            books
        WHERE
            num_pages > 50
        GROUP BY
            publisher_id
        ORDER BY
            books_cnt DESC) AS SUBQ1 ON publishers.publisher_id = SUBQ1.publisher_id
ORDER BY
    books_cnt DESC;;
'''

result3 = pd.io.sql.read_sql(query3, con = engine)

result3.head(10)
```

#### Вывод
- Самое активное издательство - `Penguin Books`.
- Разрыв в количестве изданных книг очень большой.

### Определить автора с самой высокой средней оценкой книг с 50 и более оценками

```python
query4 = ''' 
SELECT
    authors.author AS author,
    AVG(SUBQ2.avg_rating) AS avg_rating
FROM
    authors
    RIGHT JOIN (
        SELECT
            SUBQ1.title AS title,
            SUBQ1.author_id AS author_id,
            AVG(ratings.rating) AS avg_rating
        FROM
            ratings
            LEFT JOIN (
                SELECT
                    author_id AS author_id,
                    title AS title,
                    book_id AS book_id
                FROM
                    books) AS SUBQ1 ON ratings.book_id = SUBQ1.book_id
        GROUP BY
            SUBQ1.title,
            SUBQ1.author_id
        HAVING
            COUNT(rating_id) > 50) AS SUBQ2 ON authors.author_id = SUBQ2.author_id
GROUP BY
    authors.author
ORDER BY
    avg_rating DESC;
'''

result4 = pd.io.sql.read_sql(query4, con = engine)

result4
```

#### Вывод
- Совершенно не удивительно увидеть Джоан Роулинг на первом месте.

### Посчитайть среднее количество обзоров от пользователей, которые поставили больше 50 оценок

```python
query5 = ''' 
SELECT
    AVG(SUBQ2.reviews_cnt) AS avg_reviews_cnt
FROM
    (
    SELECT
        reviews.username AS username,
        SUBQ1.cnt AS ratings_cnt,
        COUNT(reviews.review_id) AS reviews_cnt
    FROM
        reviews
        RIGHT JOIN (
            SELECT
                username AS username,
                COUNT(rating_id) AS cnt
            FROM
                ratings
            GROUP BY
                username

            HAVING
                COUNT(rating_id) > 50
            ORDER BY
                cnt DESC) AS SUBQ1 on reviews.username = SUBQ1.username
    GROUP BY
        reviews.username,
        SUBQ1.cnt) AS SUBQ2;
'''

result5 = pd.io.sql.read_sql(query5, con = engine)

result5
```

#### Вывод
- Активный читатель не просто ставит много оценок, он ещё и активно обозревает прочитанное.
